```python
df['km_driven'].skew()
```

```python
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
df['km_driven_new']=np.where(df['km_driven']>IQR,IQR,df['km_driven'])
df['km_driven_new'].skew()
```

```python
#bivariate analysis of year and selling price
sns.scatterplot(data=df, x='year', y='selling_price')
plt.show()
```

```python
#13. What is the average depreciation in selling price based on the bike's age (current year - manufacturing year)
current_year = 2026
df['depreciation'] = df['ex_showroom_price'] - df['selling_price']
df['age'] = current_year -df['year']
df['depreciation_per_year'] = df['depreciation'] / df['age']
df.groupby('year')['depreciation_per_year'].mean()
```

```python
#sales vs year
plt.figure(figsize=(15,8))
sns.barplot(x='year',y='selling_price',data=df,palette='Set2')
plt.title('sales per year')
plt.xlabel('year')
plt.ylabel('Sales amount')
plt.show()
```

```python
#inside >> highest sales of the bike is in the year 2020
```
